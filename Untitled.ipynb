```python
import requests
import os
import numpy as np
import pandas as pd
from sodapy import Socrata
from pprint import pprint
import matplotlib.pyplot as plt
import scipy.stats as st
from scipy.stats import linregress
import numpy as np
```

```python
client = Socrata("data.cdc.gov", None)

results = client.get("n8mc-b4w4", limit = 150000, res_state="MN")
```

```python
results_df = pd.DataFrame.from_records(results)
results_df.head(15)
```

```python
# look for missing values
results_df.count()
```

```python
# drop null rows
no_null_results_df = results_df.dropna(how='any')
no_null_results_df.count()
```

```python
# drop NA and"Missing" rows on columns of interest.
NA_clean_null_df = no_null_results_df.loc[(no_null_results_df['race'].str.contains("NA")) == False, :]
exposure_clean_null_df = NA_clean_null_df.loc[NA_clean_null_df["exposure_yn"].str.contains("Missing") == False]
hosp_clean_null_df = exposure_clean_null_df.loc[exposure_clean_null_df["hosp_yn"].str.contains("Missing") == False]
ethnicity_clean_null_1_df = hosp_clean_null_df.loc[hosp_clean_null_df["ethnicity"].str.contains("NA") == False]
ethnicity_clean_null_2_df = ethnicity_clean_null_1_df.loc[ethnicity_clean_null_1_df["ethnicity"].str.contains("Missing") == False]
clean_null_df = ethnicity_clean_null_2_df.loc[ethnicity_clean_null_2_df["icu_yn"].str.contains("Missing") == False]
clean_null_df

#confirm the rows dropped by using .count()
```
